```python
import asyncio
import time

from langchain_openai import OpenAI
from langchain.prompts import PromptTemplate
from langchain_core.runnables import RunnableSequence


```

```python
def generate_serially():
    llm = OpenAI(temperature=0.9)
    prompt = PromptTemplate(
        input_variables=["pet"],
        template="What is a good name for a friendly pet  {pet}?",
    )
    chain =  RunnableSequence(prompt, llm)  # Create a RunnableSequence object
    for _ in range(5):
        resp = chain.invoke({"pet":"dog"}) # Invoke with a dictionary input
        print(resp)
```

```python
s = time.perf_counter()
generate_serially()
elapsed = time.perf_counter() - s
print("\033[1m" + f"Serial executed in {elapsed:0.2f} seconds." + "\033[0m")
```

```python
async def async_generate(chain):
    resp = await chain.ainvoke({"product": "icecream"})  # Pass input as a dictionary
    print(resp)

async def generate_concurrently():
    llm = OpenAI(temperature=0.9)
    prompt = PromptTemplate(
        input_variables=["product"],
        template="Give me different flavors of {product}?",
    )
    chain =  RunnableSequence(prompt, llm)
    tasks = [async_generate(chain) for _ in range(5)]
    await asyncio.gather(*tasks)


s = time.perf_counter()
await generate_concurrently()
elapsed = time.perf_counter() - s
print("\033[1m" + f"Concurrent executed in {elapsed:0.2f} seconds." + "\033[0m")
```
